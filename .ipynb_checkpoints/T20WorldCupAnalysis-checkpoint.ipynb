# T20 WORLD CUP DATA ANALYSIS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load datasets
batting_df = pd.read_csv('batting_stats.csv')
bowling_df = pd.read_csv('bowling_stats.csv')
matches_df = pd.read_csv('matches.csv')
```

```python
# Display basic information
print("Batting Data:")
print(batting_df.head())
print("\nBowling Data:")
print(bowling_df.head())
print("\nMatches Data:")
print(matches_df.head())
```

```python

# Basic statistics
print("\nBatting Stats Summary:")
print(batting_df.describe())
```

```python
# print("\nBowling Stats Summary:")
# print(bowling_df.describe())
```

# Top 10 Run Scorers

```python
# Top 10 Run Scorers
top_scorers = batting_df[['Batsman', 'Runs']].head(10)
print("\nTop 10 Run Scorers:")
print(top_scorers)
# Visualization
plt.barh(top_scorers['Batsman'], top_scorers['Runs'], color='skyblue')
plt.xlabel('Runs')
plt.ylabel('Batsman')
plt.title('Top 10 Run Scorers')
```

```python
# print("\nMatches Stats Summary:")
# print(matches_df.describe())
```

# Top 5 Wicket Takers

```python
# Top 5 Wicket Takers
top_wicket_takers = bowling_df[['Bowler', 'Wickets']].head(5)
print(top_wicket_takers)
# Visualization
plt.barh(top_wicket_takers['Bowler'],top_wicket_takers['Wickets'], color=('salmon','green', 'blue','red','purple'))
plt.xlabel('Wickets')
plt.ylabel('Bowlers')
plt.title('Top 5 Wicket Takers')
```

# Most Successful Teams

```python
# Most Successful Teams
Teams_wins = matches_df['Winning_team'].value_counts()
print(Teams_wins)
# Visualization
Teams_wins.plot(kind='bar', color='pink')
plt.xlabel('Teams')
plt.ylabel('No. Of Matches Win')
plt.title('Most Successful Teams')
```

# Venue-Based Performance

```python
# Venue-Based Performance
venue_based = matches_df[['Venue_stadium','Winning_team',]].value_counts().head(10)
print(venue_based)
# Visualization
venue_based.plot(kind='barh', color='grey')
plt.xlabel('Win Count')
plt.ylabel('Venue based teams performance')
plt.title('Venue-Based Performance')
# plt.pie(venue_based, labels=venue_based.index, autopct='%1.1f%%')
```

# Team win by No. Of Wickets or Wicket based team win

```python
# Team win by No. Of Wickets or Wicket based team win
wicket_based = matches_df[['Win_by_wickets','Winning_team',]].value_counts().head(10)
print(wicket_based)
# Visualization
wicket_based.plot(kind='barh', color='purple')
plt.xlabel('Win Count')
plt.ylabel('Team win by No. Of Wickets')
plt.title('Wicket based team win')
```

# Identifying Best Economy Bowler

```python
# Identifying Best Economy Bowler
best_economy_bowler = bowling_df.nsmallest(3, "Economy") # top 3 best economy bowler
print(best_economy_bowler)
plt.bar(best_economy_bowler["Bowler"], best_economy_bowler["Economy"], color=['green', 'blue','red'])
plt.xlabel("Bowler")
plt.ylabel("Economy Rate")
plt.title("Top 3 Most Economical Bowlers")
```

# Find Best Average

```python
# Find Best Average
best_average = batting_df.nlargest(10, "Average")
print(best_average)
plt.barh(batting_df["Batsman"], batting_df["Average"], color="red")
plt.xlabel("Average")
plt.ylabel("Batsman")
plt.title("Batting Average Analysis")
```

# Identifying best strike rate

```python
# Identifying best strike rate
best_strike_rate = batting_df.nlargest(10, "Strike_rate")
print(best_strike_rate)  
plt.barh(batting_df["Batsman"], batting_df["Strike_rate"], color="olive")
plt.xlabel("Strike Rate")
plt.ylabel("Batsman")
plt.title("Best Strike Rate  Analysis")
```

# identifying most sixes

```python
# identifying most sixes
most_sixes = batting_df.nlargest(10, "Sixes")
print(most_sixes)  
plt.barh(batting_df["Batsman"], batting_df["Sixes"], color="orange")
plt.xlabel("Most Sixes")
plt.ylabel("Batsman")
plt.title("Most Sixes Analysis")
```

# identifying most fours

```python
# identifying most fours
most_fours = batting_df.nlargest(10, "Fours")
print(most_fours)  
plt.barh(batting_df["Batsman"], batting_df["Fours"], color="green")
plt.xlabel("Most Fours")
plt.ylabel("Batsman")
plt.title("Most Fours Anlysis")
```

#  Toss Decision Analysis

```python
#  Toss Decision Analysis
toss_decision_counts = matches_df["Toss_decision"].value_counts()
print(toss_decision_counts)
plt.pie(toss_decision_counts, labels=toss_decision_counts.index, autopct='%1.1f%%', colors=['blue', 'orange'])
plt.title("Toss Decision Analysis (Bat vs Bowl)")
```

#  Most Man of the Match Awards

```python
#  Most Man of the Match Awards
mom_counts = matches_df["Man_of_match"].value_counts().head(10)
print(mom_counts)
plt.barh(mom_counts.index, mom_counts.values, color="purple")
plt.xlabel("Man of the match count")
plt.ylabel("Players")
plt.title("Most Man of the Match Awards")
```

# Super Over Matches Count

```python
#  Super Over Matches Count
super_over_count = matches_df["Super_over_match"].value_counts()
print(super_over_count)
plt.pie(super_over_count, autopct='%1.1f%%', colors=["yellow", "red"])  # %1.1f%% --> after point only 1 decimal point allow
plt.title("Super Over Matches Analysis")
```

# Matches Win By Toss Winning Team

```python
# Converts the grouped data from a multi-index format into a regular DataFrame and stored in a new column named "Wins"
 # groupby-->  we are analyzing data separately for each team that won the toss.
toss_vs_winning = matches_df.groupby("Toss_winner")["Winning_team"].value_counts().reset_index(name="Wins")
print(toss_vs_winning) 
plt.barh(toss_vs_winning["Toss_winner"], toss_vs_winning["Wins"], color="aquamarine")
plt.xlabel("Number of Wins")
plt.ylabel("Teams")
plt.title("Matches Win by Toss Winning Team")
```

```python
batting_vs_bowling = pd.merge(batting_df[["Batsman", "Runs"]],bowling_df[["Bowler", "Wickets"]],
    left_index=True, right_index=True)
print("\nAnalysis 2: Batsmen Runs vs. Bowler Wickets")
print(batting_vs_bowling)

plt.figure(figsize=(8, 5))
plt.bar(batting_vs_bowling["Batsman"], batting_vs_bowling["Runs"], color="blue", label="Runs")
plt.bar(batting_vs_bowling["Bowler"], batting_vs_bowling["Wickets"] * 10, color="red", alpha=0.7, label="Wickets x10")
plt.xlabel("Player")
plt.ylabel("Runs / Wickets (scaled)")
plt.title("Batsmen Runs vs. Bowler Wickets Performance")
plt.xticks(rotation=45,ha='right')
plt.legend()
```
